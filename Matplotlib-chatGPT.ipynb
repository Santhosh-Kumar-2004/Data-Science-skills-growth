```python
import numpy as np
import matplotlib.pyplot as plt

# Generating data for the plots
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
categories = ["A", "B", "C", "D"]
values = [3, 7, 2, 5]

# Creating a figure with 1 row and 2 columns of subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# First subplot - Line Plot
ax1.plot(x, y1, color="blue")
ax1.set_title("Line Plot")
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y-axis")

# Second subplot - Bar Plot
ax2.bar(categories, values, color="orange")
ax2.set_title("Bar Plot")
ax2.set_xlabel("Categories")
ax2.set_ylabel("Values")

# Displaying the figure with subplots
plt.show()
```

```python
# Sample data for a line plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Creating the line plot with grid lines
plt.plot(x, y, color="blue", marker="o", linestyle="--")

# Adding title and labels
plt.title("Line Plot with Grid Lines")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Adding grid lines
plt.grid(True)

# Displaying the plot
plt.show()
```

```python
# Sample data for multiple line plots
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Creating the first line plot
plt.plot(x, y1, color="blue", linestyle="--", marker="o", label="Series 1")

# Creating the second line plot
plt.plot(x, y2, color="green", linestyle="-", marker="s", label="Series 2")

# Adding title, labels, and legend
plt.title("Customized Line Plot with Different Styles")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Displaying the plot
plt.show()
```

```python
# Sample data for a line plot with two series
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Creating line plots
plt.plot(x, y1, color="blue", linestyle="--", marker="o", label="Series 1")
plt.plot(x, y2, color="green", linestyle="-", marker="s", label="Series 2")

# Adding title, labels, and legend
plt.title("Line Plot with Legend")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc="upper left")  # Position the legend in the upper-left corner

# Displaying the plot
plt.show()
```

```python
# Sample data for a scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)  # Color intensity for each point

# Creating the scatter plot with color map
plt.scatter(x, y, c=colors, cmap="viridis", s=100, edgecolor="black")

# Adding title and labels
plt.title("Scatter Plot with Color Map")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.grid(True)
# Adding color bar for reference
plt.colorbar()

# Displaying the plot
plt.show()
```

```python
# Sample data for annotation
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Creating a line plot
plt.plot(x, y, color="blue", marker="o", linestyle="--")

# Adding annotation at the point (3, 6)
plt.annotate("Important Point", xy=(3, 6), xytext=(3.5, 7),
             arrowprops=dict(facecolor="black", arrowstyle="->"))

plt.grid(True)

# Adding title and labels
plt.title("Line Plot with Annotation")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Displaying the plot
plt.show()
```

```python
# Generating random data for the histogram
data = np.random.randn(1000)

# Creating the histogram
plt.hist(data, bins=20, color="purple", alpha=0.8)

plt.grid(True)

# Adding title and labels
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Displaying the plot
plt.show()
```

```python
# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating a figure with 2 subplots arranged in a 2x1 layout
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# First subplot - sine wave
axs[0].plot(x, y1, color="blue")
axs[0].set_title("Sine Wave")
axs[0].set_xlabel("X-axis")
axs[0].set_ylabel("Y-axis")
axs[0].grid(True)

# Second subplot - cosine wave
axs[1].plot(x, y2, color="red")
axs[1].set_title("Cosine Wave")
axs[1].set_xlabel("X-axis")
axs[1].set_ylabel("Y-axis")
axs[1].grid(True)

# Adjusting layout to prevent overlap
plt.tight_layout()

# Displaying the figure with subplots
plt.show()
```

```python
import seaborn as sns

# Loading a sample dataset
tips = sns.load_dataset("tips")

# Creating a box plot using Seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set1")

# Adding title and labels
plt.title("Box Plot of Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")

# Displaying the plot
plt.show()
```

```python
# Sample data: correlation matrix of a dataset
import numpy as np
import pandas as pd

# Creating a random DataFrame
data = np.random.rand(10, 10)
df = pd.DataFrame(data, columns=[f"Feature{i}" for i in range(1, 11)])

# Creating a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Adding title
plt.title("Heatmap of Feature Correlation")

# Displaying the plot
plt.show()
```

```python
# Sample data for a pie chart
labels = ["A", "B", "C", "D"]
sizes = [40, 30, 20, 10]
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]

# Creating the pie chart
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", startangle=90)

# Adding title
plt.title("Pie Chart of Categories")

# Displaying the plot
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Sample data for a 3D scatter plot
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Creating a 3D plot
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x, y, z, c=z, cmap="viridis", s=50)

# Adding title and labels
ax.set_title("3D Scatter Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

# Displaying the plot
plt.show()
```

```python
# Sample data for a regression plot
import seaborn as sns

# Loading the 'tips' dataset
tips = sns.load_dataset("tips")

# Creating the regression plot
plt.figure(figsize=(8, 6))
sns.regplot(x="total_bill", y="tip", data=tips)

# Adding title and labels
plt.title("Regression Plot of Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Displaying the plot
plt.show()
```
