```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load a sample dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Scatter plot of total bill vs. tip
sns.scatterplot(x="total_bill", y="tip", data=tips.head())
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
# Histogram of total bills
sns.histplot(tips['total_bill'], kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

```python
# Box plot of tips by day of the week
sns.boxplot(x="day", y="tip", data=tips)
plt.title("Tip Amount by Day")
plt.show()
```

```python
# Violin plot of tips by day
sns.violinplot(x="day", y="tip", data=tips, hue="sex", split=True)
plt.title("Tip Amount by Day and Sex")
plt.show()
```

```python
# Set Seaborn style
sns.set_style("whitegrid")  # Options: white, dark, whitegrid, darkgrid, ticks
```

```python
sns.pairplot(tips, hue="sex")
plt.show()
```

# Additional plot types in seaborn

```python
sns.countplot(x="day", data=tips)
plt.title("Count of Days in Tips Dataset")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=tips, hue="sex")
plt.title("Average Total Bill by Day and Sex")
plt.show()
```

```python
# Save the plot as an image file
plt.savefig("seaborn_plot.png", dpi=300, bbox_inches="tight")
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="bar", data=tips)
plt.title("Average Total Bill by Day and Sex")
plt.show()
```

```python
g = sns.FacetGrid(tips, col="day")
g.map(plt.hist, "total_bill", bins=10)
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
plt.show()
```

```python
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("Regression Plot of Total Bill vs Tip")
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips.head(100))
plt.title("Linear Regression of Total Bill vs Tip by Sex")
plt.show()
```

```python
g = sns.PairGrid(tips, hue="sex")
g.map_diag(sns.histplot)  # Histograms on the diagonal
g.map_offdiag(sns.scatterplot)  # Scatter plots on the off-diagonals
g.add_legend()
plt.show()
```

```python
# Set a color palette
sns.set_palette("husl")  # Options include "deep", "muted", "bright", "dark", etc.

# Example with a bar plot
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill by Day with Custom Palette")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
sns.despine()  # Removes the top and right spines
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill by Day with Custom Figure Size")
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)

# Add annotations
for i, row in tips.iterrows():
    plt.text(row['total_bill'], row['tip'], f"{row['tip']:.1f}", ha="right")

plt.title("Total Bill vs Tip with Annotations")
plt.show()
```
