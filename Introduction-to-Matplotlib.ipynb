# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting.
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots()

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple plot", 
       xlabel="X-Axis", 
       ylabel="Y-Axis")

# 5. Save & show (You have the whole figure)
fig.savefig("sample-plot.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50]
y = [23, 54, 39, 99, 83]

fig, ax = plt.subplots()

ax.plot(x, y, 'r')

ax.set(title="Sample Plot",
      xlabel="This is the Xlabel",
      ylabel="This is the Ylabel")

fig.savefig("simple-plot.png")
plt.show()
```

## Making figures with NumPy arrays

```python
import numpy as np
```

```python
#Creating the data
x = np.linspace(0, 10, 100)
```

```python
x[: 10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
plt.show()
```

```python
# Use the same data for Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
plt.show()

plt.close()
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
plt.show()
```

```python
# make a plot form a dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="San's Buttert shop Prices",
      xlabel="Butter-Name",
      ylabel="Prices($)")

plt.show()
plt.close()
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
plt.show()
```

```python
# make some data for histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

# Two options for subplots

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))

#Plot to each different Axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
plt.show()
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))

plt.show()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
car_sales["Price_rupees"] = [4000, 5000, 7000, 22000, 3500, 4500, 7500, 7000, 6250, 9700]
```

```python
car_sales
```

```python
car_sales["Sale_date"] = pd.date_range("1/1/2020", periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales["Total_sales"] = car_sales["Price_rupees"].cumsum()
car_sales
```

```python
# lets plot the total sales
car_sales.plot(x="Sale_date", y="Total_sales")
plt.show()
plt.close()
```
