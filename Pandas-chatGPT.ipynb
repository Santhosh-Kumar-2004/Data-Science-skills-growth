```python
# Creating a second DataFrame for merging
other_df = pd.DataFrame({
    "ID": [1, 2, 3, 4, 5, 6],  # These IDs should match some in the original DataFrame
    "Bonus": [5000, 7000, 8000, 6000, 9000, 20000]
})

# Merging based on 'ID' column
merged_df = df.merge(other_df, on="ID", how="inner")
print(merged_df.head())
```

```python
# Creating a second DataFrame with similar columns for concatenation
new_rows = pd.DataFrame({
    "ID": range(101, 106),
    "Name": [f"Person_{i}" for i in range(101, 106)],
    "Age": np.random.randint(18, 60, size=5),
    "Country": np.random.choice(["USA", "Canada", "UK", "Germany", "France"], size=5),
    "Occupation": np.random.choice(["Engineer", "Doctor", "Artist", "Teacher", "Lawyer"], size=5),
    "Salary": np.random.randint(30000, 120000, size=5),
    "Experience": np.random.randint(1, 40, size=5),
    "Department": np.random.choice(["HR", "IT", "Finance", "Marketing", "Sales"], size=5),
    "Rating": np.random.uniform(1.0, 5.0, size=5),
    "Joining_Date": pd.to_datetime(np.random.choice(pd.date_range("2000-01-01", "2023-01-01"), size=5))
})

# Concatenating along rows
concatenated_df = pd.concat([df, new_rows], axis=0)
print(concatenated_df.tail())
```
